```python
def clasificar_cliente_edad(edad):
    if edad < 30:
        return 'Joven'
    elif edad < 60:
        return 'Adulto'
    else:
        return 'Senior'

# Aplicar funci√≥n personalizada
df['Clasificaci√≥n Edad'] = df['Age'].apply(clasificar_cliente_edad)

df # Agregamos la clasificacion de edad
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Estad√≠sticas descriptivas b√°sicas
print("Estad√≠sticas descriptivas b√°sicas:")
print(df.describe())

# Visualizaci√≥n de la distribuci√≥n de variables num√©ricas
sns.set()
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(df['Age'], ax=axs[0, 0])
sns.histplot(df['Quantity'], ax=axs[0, 1])
sns.histplot(df['Price per Unit'], ax=axs[1, 0])
sns.histplot(df['TotalAmount'], ax=axs[1, 1])
plt.tight_layout()
plt.show()

# Boxplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
sns.boxplot(df['Age'], ax=axs[0, 0])
sns.boxplot(df['Quantity'], ax=axs[0, 1])
sns.boxplot(df['Price per Unit'], ax=axs[1, 0])
sns.boxplot(df['TotalAmount'], ax=axs[1, 1])
plt.tight_layout()
plt.show()
```

```python

# Tendencias de ventas a lo largo del tiempo
df['Date'] = pd.to_datetime(df['Date'])
df['Mes'] = df['Date'].dt.month
ventas_mes = df.groupby('Mes')['TotalAmount'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=ventas_mes.index, y=ventas_mes.values)
plt.title('Ventas totales por mes')
plt.xlabel('Mes')
plt.ylabel('TotalAmount')
plt.show()
```

```python

# Relaci√≥n entre variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age'], y=df['TotalAmount'])
plt.title('Relaci√≥n entre edad y monto total')
plt.xlabel('Age')
plt.ylabel('TotalAmount')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Quantity'], y=df['TotalAmount'])
plt.title('Relaci√≥n entre cantidad y monto total')
plt.xlabel('Quantity')
plt.ylabel('TotalAmount')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Price per Unit'], y=df['TotalAmount'])
plt.title('Relaci√≥n entre precio por unidad y monto total')
plt.xlabel('Precio por unidad')
plt.ylabel('TotalAmount')
plt.show()
```

```python

# Distribuci√≥n de ventas y valores at√≠picos
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalAmount'], kde=True)
sns.boxplot(df['TotalAmount'], vert=False)
plt.title('Distribuci√≥n de ventas y valores at√≠picos')
plt.xlabel('TotalAmount')
plt.show()
```

parte 5

```python
# An√°lisis de correlaci√≥n
corr_matrix = df[['Age', 'Quantity', 'Price per Unit', 'TotalAmount']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Mapa de calor de correlaci√≥n')
plt.show()
```

```python

# Subplots para comparar variables clave
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
sns.scatterplot(x=df['Age'], y=df['TotalAmount'], ax=axs[0, 0])
sns.scatterplot(x=df['Quantity'], y=df['TotalAmount'], ax=axs[0, 1])
sns.scatterplot(x=df['Price per Unit'], y=df['TotalAmount'], ax=axs[1, 0])
sns.scatterplot(x=df['Age'], y=df['Quantity'], ax=axs[1, 1])
axs[0, 0].set_title('Edad vs. Monto total')
axs[0, 1].set_title('Cantidad vs. Monto total')
axs[1, 0].set_title('Precio por unidad vs. Monto total')
axs[1, 1].set_title('Edad vs. Cantidad')
plt.tight_layout()
plt.show()
```

```python

# Gr√°fico de dispersi√≥n con cuadr√≠cula y leyenda
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age'], y=df['TotalAmount'])
plt.grid(True)
plt.legend(labels=['Datos'])
plt.title('Relaci√≥n entre edad y monto total')
plt.xlabel('Edad')
plt.ylabel('Monto total')
plt.show()
```

```python

# Gr√°fico de barras apiladas con anotaciones
plt.figure(figsize=(10, 6))
sns.barplot(x=df['ProductCategory'], y=df['TotalAmount'])
plt.title('Monto total por categor√≠a de producto')
plt.xlabel('Categor√≠a de producto')
plt.ylabel('Monto total')
for i, patch in enumerate(plt.gca().patches):
    plt.gca().text(patch.get_x() + patch.get_width()/2, patch.get_height(),
                   str(round(patch.get_height(), 2)), ha='center', va='bottom')
plt.show()
```

```python

# Gr√°fico de l√≠neas con flechas
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Date'], y=df['TotalAmount'])
plt.title('Ventas totales por fecha')
plt.xlabel('Fecha')
plt.ylabel('Monto total')
plt.gca().arrow(0.5, 0.5, 0.1, 0.1, head_width=0.05, head_length=0.1, fc='k', ec='k')
plt.show()
```
